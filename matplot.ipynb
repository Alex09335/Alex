```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,10,13]
plt.plot(x,y,color="red",marker="o")
plt.title("example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=["ali","reza","arash","erfan"]
y=[1000,2000,1500,5000]
plt.bar(x,y,color="red")
plt.title("example")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
